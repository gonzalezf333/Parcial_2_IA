## 1.1 Cambio de formato de notas finales 
las notas inicialmente se encuentran con el siguente formato: "1F-1,2F-1"
utilice una funcion que recorre la cadena y retorna el promedio de la misma
si esta es no vacia, caso contrario le asigna 0 a la misma

```python
import statistics
def cambiar_nota(a): #funcion que retorna el promedio de las notas
    notas=[]
    j=""
    for i in a:
        if(j=="-"):
            if(i=="C" or  i=="D"):
                pass
            else:
                notas.append(int(float(i)))
        j=i
    mean = statistics.mean(notas)
    return mean
```

```python
MCT_CB = mct_cb.copy(deep=True)
```

```python
MCT_CB[['Nota.Final']] = MCT_CB[['Nota.Final']].fillna('f-0')
MCT_CB['Nota.Final'] = MCT_CB['Nota.Final'].apply(cambiar_nota)
MCT_CB
```

```python
MCT_CB.to_csv('cb_mct.csv', index=False)
```
